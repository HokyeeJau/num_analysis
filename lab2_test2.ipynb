# Declare a Python list

```python
hilbert_matrix = []

for i in range(1, 101):
    row = []
    for j in range(1, 101):
        row.append(1/(i+j-1))
    hilbert_matrix.append(row)
```

# Let list into numpy.ndarray and show its shape(dimension)

```python
import numpy as np
np.set_printoptions(threshold=100000000000)
# Show all the data in the matrix

hilbert_matrix = np.array(hilbert_matrix)

hilbert_matrix.shape
```

# Check if the matrix runs well

```python
hilbert_matrix[0]
```

# Show out the inverse matrix of Hilbert Matrix

```python
np.matrix(hilbert_matrix).I
```

# The following is the same as the above.(N = 200, 500)
